```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train,y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
#scaling the values to improve accuracy
X_train = X_train / 255

X_test = X_test / 255
```

```python
X_train[0]
```

flattening the 28X28 grid into a 1d array

```python
 X_train_flattened = X_train.reshape(len(X_train),28*28)
 X_test_flattened = X_test.reshape(len(X_test),28*28)
```

```python
X_train_flattened[0]
```

```python
y_train
```

```python
model =keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])

model.compile(
      optimizer="adam" ,
      loss="sparse_categorical_crossentropy",
      metrics=["accuracy"]
)

model.fit(X_train_flattened,y_train,epochs=5)
```

```python
model.evaluate(X_test_flattened , y_test)
```

```python
plt.matshow(X_test[0])
```

sample prediction

```python
predicted_values =  model.predict(X_test_flattened)

predicted_values[0 ] #prints the scores from 0 to 9
```

```python
np.argmax(predicted_values[0])
```

```python
y_predicted_labels = [np.argmax(i) for i in predicted_values]
y_predicted_labels[:5]
```

```python
#including a confusion matrix

cm = tf.math.confusion_matrix(labels = y_test , predictions = y_predicted_labels )
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#adding a hidden layer to the simple neural network

model =keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
    ])

model.compile(
      optimizer="adam" ,
      loss="sparse_categorical_crossentropy",
      metrics=["accuracy"]
)

model.fit(X_train_flattened,y_train,epochs=5)
```

```python
model.evaluate(X_test_flattened , y_test)
```

```python
predicted_values = model.predict(X_test_flattened)
y_predicted_labels = [np.argmax(i) for i in predicted_values]
#including a confusion matrix
cm = tf.math.confusion_matrix(labels = y_test , predictions = y_predicted_labels )

#plotting the heatmap
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#adding a hidden layer to the simple neural network

model =keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)), #alternate for not manually making hte flat array

    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
    ])

model.compile(
      optimizer="adam" ,
      loss="sparse_categorical_crossentropy",
      metrics=["accuracy"]
)

model.fit(X_train,y_train,epochs=5)
```
